1 Bag of Words (BoW) / TF-IDF

Input to model:

A list of documents

Each document is a row of raw text (string)

Example:

[
  "good phone",
  "battery life is good"
]


Preprocessing:

Tokenization  handled internally by the vectorizer

Stop-word removal  Yes (commonly used)

Lemmatization  Optional

Stemming  Optional

Best practice:

Pass raw text to CountVectorizer / TfidfVectorizer

Avoid manual tokenization unless required

2 Word2Vec

Input to model:

A list of tokenized sentences

Each sentence is a list of words

Example:

[
  ["good", "phone"],
  ["battery", "life", "good"]
]


Preprocessing:

Tokenization  Mandatory

Stop-word removal  Recommended

Lemmatization  Recommended

Stemming  Optional

Best practice:

Preprocess text before training Word2Vec

Preserve sentence boundaries

3 Deep Learning Models (LSTM, BERT, etc.)

Input to model:

Raw text (string)

Preprocessing:

Tokenization  handled by model tokenizer

Stop-word removal  Not recommended

Lemmatization  Not recommended

 Internal Working Summary: BoW, TF-IDF & Word2Vec
 Bag of Words (BoW)

BoW first builds a common vocabulary from all training documents.
Each document is then represented as a fixed-length vector where each value is the count of a word from the vocabulary in that document.

 TF-IDF

TF-IDF also builds a common vocabulary like BoW.
Each document is converted into a vector, but words are weighted based on how frequent they are in the document (TF) and how rare they are across all documents (IDF).

 Word2Vec

Word2Vec builds one global vocabulary and processes text sentence by sentence.
Each word is internally represented as a one-hot vector, and a sliding context window is applied within each sentence to train a neural network that learns dense word embeddings.

Example Text
Doc1: "good phone"
Doc2: "battery life good"

 Bag of Words (BoW)

Vocabulary (from all documents):

['battery', 'good', 'life', 'phone']


Vectors:

Doc1  [0, 1, 0, 1]
Doc2  [1, 1, 1, 0]


 Each value = word count in the document

 TF-IDF

Same vocabulary:

['battery', 'good', 'life', 'phone']


Vectors (weighted):

Doc1  [0.0, 0.3, 0.0, 0.7]
Doc2  [0.6, 0.2, 0.6, 0.0]


 Common words (like good) get lower weight, rare words get higher weight

##AvgWord2Vec

```python
def avg_word2vec(sent,model,lemmatizer,stopwords_set):
    words = simple_preprocess(sent)
    # Apply stopword removal and lemmatization consistently with corups_tock creation
    filtered_words = [word for word in words if word not in stopwords_set]
    processed_words = [lemmatizer.lemmatize(word) for word in filtered_words]

    vectors = [model[word] for word in processed_words if word in model.key_to_index]

    if len(vectors) == 0:
        return np.zeros(model.vector_size)

    return np.mean(vectors, axis=0)
```

```python
X_train_w2v = np.concatenate((np.array([avg_word2vec(sent,w2v.wv,lem,stopwords) for sent in X_train['text']]),np.array([avg_word2vec(sent,w2v.wv,lem,stopwords) for sent in X_train['title']])),axis=1)
X_train_w2v = np.concatenate((X_train_w2v,X_train.drop(columns=['text','title']).to_numpy()),axis=1)
```

```python
X_test_w2v = np.concatenate((np.array([avg_word2vec(sent,w2v.wv,lem,stopwords) for sent in X_test['text']]),np.array([avg_word2vec(sent,w2v.wv,lem,stopwords) for sent in X_test['title']])),axis=1)
X_test_w2v = np.concatenate((X_test_w2v,X_test.drop(columns=['text','title']).to_numpy()),axis=1)
```

 Multinomial Naive Bayes (MultinomialNB)

 Used mainly for text classification

 Works with non-negative, count-based features

 Best with Bag of Words and TF-IDF

 Common tasks:

Spam detection

Sentiment analysis

Topic classification

 Very fast and memory-efficient

 Not suitable for Word2Vec or other continuous embeddings, because it gives negative value feature

```python
model_w2v = LogisticRegression(max_iter=1000)
model_w2v.fit(X_train_w2v,y_train)
y_w2v_pred = model_w2v.predict(X_test_w2v)
print(accuracy_score(y_w2v_pred,y_test))
```
