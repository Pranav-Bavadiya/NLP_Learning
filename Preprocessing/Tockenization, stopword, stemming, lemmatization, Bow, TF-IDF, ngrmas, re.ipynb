```python
import nltk
```

```python
sentences = []
for reviwe in df['review_text'].values:
  sentences.append(reviwe)
```

```python
print(sentences[0])
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stwords = set(stopwords.words('english'))
```

```python
corups=[]
for sentense in sentences:
  new_sentences = sent_tokenize(sentense)
  for new_sentense in new_sentences:
    new_words = word_tokenize(new_sentense)
    for new_word in new_words:
      if new_word not in stwords:
        corups.append(new_word)
```

```python
corups[0]
len(corups)
```

```python
import re
```

| Regex    | Meaning                 |
| -------- | ----------------------- |
| `\w`     | letters + numbers + `_` |
| `\W`     | everything except that  |
| `\d`     | digits                  |
| `\s`     | spaces, tabs, newlines  |
| `[^a-z]` | NOT a to z              |

Whatever the regex MATCHES ‚Üí gets REPLACED / REMOVED

```python
for i in range(len(corups)):
  corups[i] = re.sub(r'\W', ' ', corups[i]).lower()
```

```python
corups[0]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
corups_stemmed = []
for word in corups:
  corups_stemmed.append(stemmer.stem(word))
```

```python
corups_stemmed[0]
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
corups_lemmatized = []
for word in corups:
  corups_lemmatized.append(lemmatizer.lemmatize(word))
```

```python
corups_lemmatized[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True, ngram_range=(1,2))
```

```python
X = cv.fit_transform(corups_lemmatized)
```
